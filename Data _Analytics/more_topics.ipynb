### List comprehension

```python
data_olympic = pd.read_csv('./athlete_events.csv')
```

```python
data_olympic.head()
```

```python
data_olympic['Gold'] = [1 if medal == 'Gold' else 0 for medal in data_olympic.Medal]
```

```python
data_olympic
```

```python
data_olympic['Silver'] = [1 if medal == 'Silver' else 0 for medal in data_olympic. Medal]
```

```python
data_olympic['Bronze'] = [1 if medal == 'Bronze' else 0 for medal in data_olympic.Medal]
```

```python
data_olympic['Silver']
```

```python
data_olympic
```

```python
data_olympic['Total'] = data_olympic.Gold + data_olympic.Silver + data_olympic.Bronze
```

```python
data_olympic
```

** to find which team got maximum number of medals

```python
data_olympic.groupby(['Team'])['Total'].sum()
```

```python
data_olympic.groupby(['Team'])['Total'].sum().reset_index()
```

```python
data_medal_count = data_olympic.groupby(['Team'])['Total'].sum().reset_index(name = 'TOTAL')
```

```python
data_medal_count
```

** To get first 10 team, who got maximum number of medals

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = data_medal_count.nlargest(10, 'TOTAL'), x = 'Team', y = 'TOTAL')
plt.show()
```

```python
data_Gold = data_olympic.groupby(['Team'])['Gold'].sum().reset_index()
```

```python
data_Gold
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = data_Gold.nlargest(10,'Gold'), x='Team', y='Gold')
plt.show()
```

```python
data_silver = data_olympic.groupby(['Team'])['Silver'].sum().reset_index()
data_silver
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = data_silver.nlargest(10,'Silver'), x='Team', y='Silver')
plt.show()
```

```python
data_Bronze = data_olympic.groupby(['Team'])['Bronze'].sum().reset_index()
data_Bronze
```

```python
plt.figure(figsize =(15,6))
sns.barplot(data = data_Bronze.nlargest(10,'Bronze'), x ='Team', y='Bronze')
plt.show()
```

* to find total points(Gold =5, Silver =3, Bronze = 1)

```python
data_olympic['Gold_point'] = [5 if medal == 'Gold' else 0 for medal in data_olympic.Medal]
```

```python
data_olympic['Silver_point'] = [3 if medal == 'Silver' else 0 for medal in data_olympic.Medal]
```

```python
data_olympic['Bronze_point'] = [1 if medal == 'Bronze' else 0 for medal in data_olympic.Medal]
```

```python
data_olympic
```

```python
data_olympic['Total_points'] = data_olympic['Gold_point'] + data_olympic['Silver_point'] + data_olympic['Bronze_point']
```

```python
data_olympic['Total_points']
```

```python
data_olympic
```

```python
data_medal_point = data_olympic.groupby(['Team'])['Total_points'].sum().reset_index()
```

```python
data_medal_point
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data =data_medal_point.nlargest(10,'Total_points'), x='Team', y='Total_points')
plt.show()
```

```python
data_gold_points = data_olympic.groupby(['Team'])['Gold_point'].sum().reset_index()
data_gold_points
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = data_gold_points.nlargest(10,'Gold_point'), x= 'Team', y= 'Gold_point')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data =data_silver_points.nlargest(10,'Silver_point'), x='Team', y='Silver_point')
plt.show()
```
