#### 27) which team got maximum number of medals?

```python
data['Gold'] = [1 if medal == 'Gold' else 0 for medal in data.Medal]
```

```python
data['Silver'] = [1 if medal == 'Silver' else 0 for medal in data.Medal]
```

```python
data['Bronze'] = [1 if medal == 'Bronze' else 0 for medal in data.Medal]
```

```python
data
```

```python
data['Total'] = data.Gold + data.Silver + data.Bronze
```

```python
data
```

```python
data_medal_count = data.groupby(['Team'])['Total'].sum().reset_index(name = 'TOTAL')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize =(16,8))
sns.barplot(data = data_medal_count.nlargest(10,'TOTAL'),x='Team',y='TOTAL')
```

### 28) which team got maximum number of gold medals?
