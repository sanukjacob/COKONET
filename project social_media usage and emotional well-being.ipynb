# SOCIAL_MEDIA USAGE AND EMOTIONAL WELL_BEING

# INTRODUCTION

Social media is a digital technology that allows sharing of ideas and information including text and visuals, through virtual networks and communications. However, multiple studies have found a strong link between heavy social media use and an increased risk for anxiety,anger and sadness.
This analysis explores a dataset that captures social media usage and users dominant emotional states. The goal of of this Project is to understand the relationship between social media usage patterns and emotional well-being.

# ABOUT THE DATASET

This unique dataset was meticulously researched and prepared by AI inventor Emirhan Bulut. It captures valuable information on social media usage and the dominant emotional state of users based on their activities. The dataset is ideal for exploring the relationship between social media usage patterns and emotional well-being.
The dataset contain 10 columns and 103 rows.

Feature description:

<li>User_ID:   Unique identifier for the user.
<li>Age:   Age of the user.
<li>Gender:   Gender of the user (Female, Male, Non-binary).
<li>Platform:   Social media platform used (e.g., Instagram, Twitter, Facebook, LinkedIn, Snapchat, Whatsapp, Telegram).
<li>Daily_Usage_Time (minutes):   Daily time spent on the platform in minutes.
<li>Posts_Per_Day:   Number of posts made per day.
<li>Likes_Received_Per_Day:   Number of likes received per day.
<li>Comments_Received_Per_Day:   Number of comments received per day.
<li>Messages_Sent_Per_Day:   Number of messages sent per day.
<li>Dominant_Emotion:   User's dominant emotional state during the day (e.g., Happiness, Sadness, Anger, Anxiety, Boredom, Neutral).
